## Цикл for

## range() - диапазон
Функция `range()` представляет собой неизменяемую последовательность целых чисел. Можно условно представить в виде списка или кортежа, но не хранящегося целиком в памяти, а вычисляющего новое значение по мере необходимости.

https://docs.python.org/3/library/functions.html#func-range

````python
range(stop)
range(start, stop, [step])
````

Обратите внимание, что числа генерируются начиная со `start` включительно, заканчивая `stop`, но не включая его:

range(stop)
range(start, stop[, step])
Параметры функции:

start - с какого числа начинается последовательность. По умолчанию - 0
stop - до какого числа продолжается последовательность чисел. Указанное число не включается в диапазон
step - с каким шагом растут числа. По умолчанию 1
Функция range хранит только информацию о значениях start, stop и step и вычисляет значения по мере необходимости. Это значит, что независимо от размера диапазона, который описывает функция range, она всегда будет занимать фиксированный объем памяти.

````python
range(10)
````

````python
list(range(10)) # Представили в виде списка
````

````python
for i in range(10,0,-1):
    print(i, end=' ')
````

**Пример 1**. Распечатать арифметическую прогрессию с первым членом 5, последним $-$ 40 и разностью прогрессии 5.

````python
for i in range(5,41,5):
    print(i, end=' ')
````

**Пример 2**. Распечатать $n$ первых членов целочисленной арифметической прогрессии с первым членом $a_1$ и разностью прогрессии $d\ne0$.

````python
# Арифметическая прогрессия
a_1 = int(input())
d = int(input())
n = int(input())
for i in range(a_1, a_1 + n * d, d):
    print(i, end=' ')
````

**Пример 3**. Распечатать $n, n>0$ первых членов целочисленной арифметической прогрессии с первым членом $a_1$ и разностью прогрессии $d\ne0$ в виде $a_1 + a_2 + ... + a_n = S$, где $S -$ сумма последовательности.

````python
# Арифметическая прогрессия
a_1, d, n = int(input()), int(input()), int(input())
a_n = a_1 + (n-1)*d
summa = int((a_1 + a_n)/2*n)
for i in range(a_1, a_n, d):
    print(i, end=' + ')
print(a_n,'=', summa)
````

````python
# решение имеющимися средствами для нормальных людей
a_1, d, n = int(input()), int(input()), int(input())
summa = 0
current = a_1
for _ in range(n - 1):
    summa += current
    print(current, end=' + ')
    current += d
print(current, '=', summa + current)
````

````python
# это светлое будущее
a_1, d, n = int(input()), int(input()), int(input())
progression = range(a_1, a_1 + d*n, d)
print(' + '.join(map(str, progression)) + ' = ' + str(sum(progression)))
````

**Пример 4**. Распечатать $n$ первых членов арифметической прогрессии (не обязательно целочисленной) с первым членом $a_1$ и разностью прогрессии $d\ne0$.

````python
a_1, d, n = float(input()), float(input()), int(input())
current = a_1
for _ in range(n):
    print(current, end=' ')
    current += d
````

**Последовательность**. Дано число $n$. Вывести последовательность чисел $1, \frac{1 }{ 2},\frac{1 }{ 3}, \dots,\frac{1 }{ n}$

````python
Variable = int(input('Введите число.'));
Sum = 0;
for Counter in range(1, Variable + 1, 1):
    print(1 / Counter);
````

**Что-то не сходится...**. Дано число $n$. Найти частичную сумму ряда $1 + \frac{1 }{ 2} + \frac{1 }{ 3} + \dots + \frac{1 }{ n}$

````python
Variable = int(input('Введите число.\n'));
Sum = 0;
for Counter in range(1, Variable + 1, 1):
    Sum += (1 / Counter);
print(Sum);
````

**Знакочередующийся**. Дано число $n$. Найти частичную сумму ряда $1 - \frac{1 }{ 2} + \frac{1 }{ 3} - \dots + \frac{(-1)^{(n+1)} }{ n}$

````python
Variable = int(input('Введите число.'));
Sum = 0;
for Counter in range(1, Variable + 1, 1):
    Sum += (pow(-1, Counter + 1) / Counter);
print(Sum);
````

````python
import math;
Variable_Start = int(input('Введите начальное значение отрезка.'));
Variable_End = int(input('Введите конечное значение отрезка.'));
Step = int(input('Введите значение шага.'));
print('X\tf(x)');
for Counter in range(Variable_Start, Variable_End + Step, Step):
    print(str(Counter) + '\t' + str(Counter ** 2 - math.sin(Counter)));
````

**Пифагоровы тройки**. Найдите количество прямоугольных треугольников с целочисленными сторонами, меньшими 100.
